```python
%load_ext autoreload
%autoreload 2
from harambot.lib import analyze
import os
import datetime

DATA_PATH = os.path.join('harambot','data','fake.json')
dt = analyze.read(DATA_PATH,keep_event=False)
first_message = analyze.get_first_message(dt)
first_message.head(n=3)
```

```python
# Find message with reactions
dt['msgWith'] = dt['messageReactions'].apply(lambda k: any(k))
# Get the number of reactions
dt['messageReactions'] = dt['messageReactions'].apply(lambda k: len(k))

reac_per_sender = dt.groupby('senderID')['messageReactions']

dt = (dt.groupby('senderID')['msgWith'].sum() / reac_per_sender.count()).reset_index()
dt = dt.merge(reac_per_sender.mean().reset_index(), on='senderID')

# Rename columns
dt.columns = ['senderID', 'msg with emote', 'emote per msg']
```
